```python
indianapolis_zipcodes = ["46107", "46217", "46218", "46219", "46220","46221", "46222", "46224", "46225", "46226",
                        "46227", "46228", "46229", "46231", "46234", "46235", "46201", "46202", "46203", "46204",
                        "46205", "46208", "46214", "46237", "46239", "46240", "46241", "46250", "46254", "46256",
                        "46259", "46260", "46268", "46278", "46216", "46236", "46183"]
```

```python
indianapolis_zipcodes_dataavailable = ["46107", "46217", "46218", "46219", "46220"]
```

```python
Indianapolis_Housing_Data = pd.DataFrame()
```

```python
for zip in indianapolis_zipcodes:
    dataframe_list = []
    
    try:
        #One Bedroom DataFrame
        one_bedroom_indicator = f"Zillow/Z{zip}_ZHVI1B"
        one_bedroom = quandl.get(one_bedroom_indicator)
        one_bedroom_df = pd.DataFrame(one_bedroom)
        one_bedroom_final = one_bedroom_df.rename(columns={"Value": "Median Price 1 Bedroom"})
        dataframe_list.append(one_bedroom_final)
    except:
        print(one_bedroom_indicator)
    
    try:
        #Two Bedroom DataFrame
        two_bedroom_indicator = f"Zillow/Z{zip}_ZHVI2B"
        two_bedroom = quandl.get(two_bedroom_indicator)
        two_bedroom_df = pd.DataFrame(two_bedroom)
        two_bedroom_final = two_bedroom_df.rename(columns={"Value": "Median Price 2 Bedroom"})
        dataframe_list.append(two_bedroom_final)
    except:
        print(two_bedroom_indicator)
        
    try:
        #Three Bedroom DataFrame
        three_bedroom_indicator = f"Zillow/Z{zip}_ZHVI3B"
        three_bedroom = quandl.get(three_bedroom_indicator)
        three_bedroom_df = pd.DataFrame(three_bedroom)
        three_bedroom_final = three_bedroom_df.rename(columns={"Value": "Median Price 3 Bedroom"})
        dataframe_list.append(three_bedroom_final)
    except:
        print(three_bedroom_indicator)

    try:
        #Four Bedroom DataFrame
        four_bedroom_indicator = f"Zillow/Z{zip}_ZHVI4B"
        four_bedroom = quandl.get(four_bedroom_indicator)
        four_bedroom_df = pd.DataFrame(four_bedroom)
        four_bedroom_final = four_bedroom_df.rename(columns={"Value": "Median Price 4 Bedroom"})
        dataframe_list.append(four_bedroom_final)
    except:
        print(four_bedroom_indicator)
        
    try:
        #Five Bedroom DataFrame
        five_bedroom_indicator = f"Zillow/Z{zip}_ZHVI5B"
        five_bedroom = quandl.get(five_bedroom_indicator)
        five_bedroom_df = pd.DataFrame(five_bedroom)
        five_bedroom_final = five_bedroom_df.rename(columns={"Value": "Median Price 5 Bedroom"})
        dataframe_list.append(five_bedroom_final)
    except:
        print(five_bedroom_indicator)

    try:
        #Median Market Value (Single Family)
        median_market_value_indicator = f"Zillow/Z{zip}_ZHVISF"
        median_market_value = quandl.get(median_market_value_indicator)
        median_market_value_df = pd.DataFrame(median_market_value)
        median_market_value_final = median_market_value_df.rename(columns={"Value":"Median Market Value (Single Family)"})
        dataframe_list.append(median_market_value_final)
    except:
        print(median_market_value_indicator)
        
    try:
        #Sales per Month
        salesPerMonth_indicator = f"Zillow/Z{zip}_SALES"
        salesPerMonth = quandl.get(salesPerMonth_indicator)
        salesPerMonth_df = pd.DataFrame(salesPerMonth)
        salesPerMonth_final = salesPerMonth_df.rename(columns={"Value":"Sales per Month"})
        dataframe_list.append(salesPerMonth_final)
    except:
        print(salesPerMonth_indicator)

    try:
        #Median List Price
        medianListPrice_indicator = f"Zillow/Z{zip}_MLPSF"
        medianListPrice = quandl.get(medianListPrice_indicator)
        medianListPrice_df = pd.DataFrame(medianListPrice)
        medianListPrice_final = medianListPrice_df.rename(columns={"Value":"Median List Price"})
        dataframe_list.append(medianListPrice_final)
    except:
        print(medianListPrice_indicator)

    try:
        #Median List Price per Square Foot
        medianPrice_squareFoot_indicator = f"Zillow/Z{zip}_MLPFSF"
        medianPrice_squareFoot = quandl.get(medianPrice_squareFoot_indicator)
        medianPrice_squareFoot_df = pd.DataFrame(medianPrice_squareFoot)
        medianPrice_squareFoot_final = medianPrice_squareFoot_df.rename(columns={"Value":"Median List Price per Square Foot"})
        dataframe_list.append(medianPrice_squareFoot_final)
    except:
        print(medianPrice_squareFoot_indicator)

    try:
        #Median % Reduction
        percentReduction_indicator = f"Zillow/Z{zip}_MPPRSF"
        percentReduction = quandl.get(percentReduction_indicator)
        percentReduction_df = pd.DataFrame(percentReduction)
        percentReduction_final = percentReduction_df.rename(columns={"Value":"Median % Reduction"})
        dataframe_list.append(percentReduction_final)
    except:
        print(percentReduction_indicator)

    try:
        #Percentage of Homes Increasing in Value
        percentIncreasing_indicator = f"Zillow/Z{zip}_PHDVAH"
        percentIncreasing = quandl.get(percentIncreasing_indicator)
        percentIncreasing_df = pd.DataFrame(percentIncreasing)
        percentIncreasing_final = percentIncreasing_df.rename(columns={"Value":"% of Homes Increasing in Value"})
        dataframe_list.append(percentIncreasing_final)
    except:
        print(percentIncreasing_indicator)

    try:
        #Percentage of Homes Decreasing in Value
        percentDecreasing_indicator = f"Zillow/Z{zip}_PHIVAH"
        percentDecreasing= quandl.get(percentDecreasing_indicator)
        percentDecreasing_df = pd.DataFrame(percentDecreasing)
        percentDecreasing_final = percentDecreasing_df.rename(columns={"Value":"% of Homes Decreasing in Value"})
        dataframe_list.append(percentDecreasing_final)
    except:
        print(percentDecreasing_indicator)

    try:
        #Percentage of Listings with Price Reduction
        percentListedReduction_indicator = f"Zillow/Z{zip}_PLPRSF"
        percentListedReduction= quandl.get(percentListedReduction_indicator)
        percentListedReduction_df = pd.DataFrame(percentListedReduction)
        percentListedReduction_final = percentListedReduction_df.rename(columns={"Value":"% of Listings with Price Reduction"})
        dataframe_list.append(percentListedReduction_final)
    except:
        print(percentListedReduction_indicator)
    
    #Merge DataFrames
    from functools import reduce
    zipData = reduce(lambda left,right: pd.merge(left,right,on='Date', how='outer'), dataframe_list)
    zipData["Zip_Code"] = zip
    zipData.sort_values(by=['Date'], inplace=True)
    
    
    #Add DataFrame for each Zip Code to the final DataFrame
    Indianapolis_Housing_Data = pd.concat([Indianapolis_Housing_Data, zipData])
    
    
    
    
    
    
```
